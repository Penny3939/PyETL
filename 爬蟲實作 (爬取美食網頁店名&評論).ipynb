# 抓取網頁上搜尋結果的內容(頁面:1)

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://tabelog.com/tw/tokyo/rstLst/"
request = urlopen(url)
html = BeautifulSoup(request)


for a in html.findAll("li",class_="list-rst"):  #可寫成("li",{"class":"list-rst"})
    en = a.find("a",class_="list-rst__name-main") #取英文店名
    ch = a.find("small",class_="list-rst__name-ja") #取中文店名
    rating = a.find("b",class_ = "c-rating__val") #取rating幾顆星
    money = a.find_all("span",class_="c-rating__val") #取價位


    #取內容.text
    print(en.text, en["href"]) #取英文店名 & 取href後面接的網址
    print(ch.text)             #取中文店名
    print(rating.text)         #取rating幾顆星
    print(money[0].text)       #取價位第一個(夜間)
    print(money[1].text)       #取價位第二個(白天)
    print("========")          #====分隔線===
```

# 抓取網頁上搜尋結果的內容(頁面:1~最後一頁),最後一頁會print"last_page"

```python

# 加上While True, try except

from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

page =50    #設定從50頁開始,不然會跑太久
while True: #不固定次數,因不知道最後一頁是多少

    url = "https://tabelog.com/tw/tokyo/rstLst/" + str(page) +"/"
    try:
        request = urlopen(url)
        
    except HTTPError:
        print("最後一頁")
        break
    html = BeautifulSoup(request)


    for a in html.findAll("li",class_="list-rst"):  #可寫成("li",{"class":"list-rst"})
        en = a.find("a",class_="list-rst__name-main") #取英文店名
        ch = a.find("small",class_="list-rst__name-ja") #取中文店名
        rating = a.find("b",class_ = "c-rating__val") #取rating幾顆星
        money = a.find_all("span",class_="c-rating__val") #取價位


        #取內容.text
        print(en.text, en["href"]) #取英文店名 & 取href後面接的網址
        print(ch.text)             #取中文店名
        print(rating.text)         #取rating幾顆星
        print(money[0].text)       #取價位第一個(夜間)
        print(money[1].text)       #取價位第二個(白天)
        print("========")          #====分隔線===

    page+=1
```
