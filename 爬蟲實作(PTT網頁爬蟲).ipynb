# 爬取PTT網頁(Get/ Post)

### 1.爬取PTT文章 (Get)
>
> 步驟一: 先將整篇文章爬取下來  
> 步驟二: 將要的資料留下,不要的刪除(利用extract)  
> 步驟三: 尋找關鍵詞,利用Jieba套件  

```python
#步驟一: 先將整篇文章爬取下來 

from urllib.request import urlopen,Request
from bs4 import BeautifulSoup
import requests
from jieba.analyse import extract_tags

url = "https://www.ptt.cc/bbs/movie/M.1684208664.A.727.html"

response = requests.get(url)
html = BeautifulSoup(response.text)
print(html)


#============================================================
#方法二: (和上方結果是一樣的)

# from urllib.request import urlopen,Request
# from bs4 import BeautifulSoup
# url = "https://www.ptt.cc/bbs/movie/M.1684208664.A.727.html"

# r= Request(url)
# r.add_header("User-Agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36")
# response = urlopen(r)

# html = BeautifulSoup(response)
# print(html)
```

### 2.留下要的資料

```python
#步驟二:留下要的資料

content = html.find("div",id= "main-content")
# print(content.text)

metas = html.find_all("span", class_="article-meta-value")
# print(metas)
print("ID:", metas[0].text)
print("看板:", metas[1].text)
print("標題:", metas[2].text)
print("時間:", metas[3].text)

#extract 刪除,人間蒸發
ms = content.find_all("div", class_= "article-metaline")          #移除作者字樣
for m in ms:
    m.extract()

ms = content.find_all("div", class_= "article-metaline-right")    #移除看板 & movie字樣
for m in ms:
    m.extract()
    
push = content.find_all("div", class_= "push")                   #找出推跟噓的留言,計算分數:推+1,噓-1,箭頭0
score = 0
for p in push:
    tag = p.find("span", class_="push-tag").text                 #.text只取內容文字(推/噓/->)
    if "推" in tag:
        score +=1
    elif "噓" in tag:
        score-=1
    p.extract()    #計算好留言的分數後,將留言刪除

print("分數:", score)
print("本文內容:", content.text)
print("關鍵詞:", extract_tags(content.text,10))                 #利用jieba套件,尋找本文關鍵詞,設定10個
```

### 1.爬取PTT文章 (Post)  --> 跟Get的差別在爬取文章時,需要多加cookies變數
