# 爬取文章裡的圖片至本地端資料夾

```python
import os.path
import pprint #讓版面比較好觀看
import json
import requests

#檢查目前位置底下有無此資料夾,若無則新增
folder_path = "./photo"
if not os.path.exists(folder_path):
    os.mkdir(folder_path)


#url=要爬的網址
url="https://www.nownews.com/nn-client/api/v1/cat/column/?pid=6085672"

headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
}

res = requests.get(url,headers = headers)

# print (res.text)
#
# json_data = json.loads(res.text)
# pprint.pprint(json_data)
# #上方兩行程式碼可縮減成以下
# pprint.pprint(res.json())

#要爬內文的標題title & 網址url & 圖片image
for article_obj in res.json()["data"]["newsList"]:
    post_title = article_obj["postTitle"]
    post_url = "https://www.nownews.com/" + article_obj["postUrl"]
    image_url = article_obj["imageUrl"]
    image_concent = requests.get(image_url, headers=headers).content

    #把圖片抓取至新資料夾photo
    image_type = image_url.split('.')[-1].split('?')[0].replace('/','-')
    image_path = f"{folder_path}/{post_title}.{image_type}"
    #可縮減成下方一行
    # image_path = f"{folder_path}/{post_title}.{image_url.split('-')[-1].split('?')[0].replace('/','-')}"

    with open(image_path, "wb") as f :
        f.write(image_concent)

    print(post_title)
    print(post_url)
    print(image_url)
    print(image_path)
    print("==========")
```

```python
#要爬的網頁是POST (無法直接copy url顯示內文)
import pprint
import requests
from bs4 import BeautifulSoup

url="https://www.newmobilelife.com/wp-json/csco/v1/more-posts"

headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
}

#取formdata: key & value值
data_str = """action: csco_ajax_load_more
page: 2
posts_per_page: 30"""
#"""不能換行,不然會出錯

#把string 轉成dictionary{}
data = {row.split(": ")[0]: row.split(": ")[1] for row in data_str.split("\n")}
#print(data)

res = requests.post(url, headers = headers, data = data)
json_data = res.json()
"""
json_data:
{'data': {'content': '\n'
                     '\n'
                     '<article class="post-482716 post type-post '
                     'status-publish format-standard has-post-thumbnail '
"""
# pprint.pprint(json_data)

#取出HTML的字串
html_str= json_data["data"]["content"]

#轉成json的字串
soup = BeautifulSoup(html_str, "html.parser")
# print(soup)

#print出標題title(從網頁element抓取h2 class值)
for title_obj in soup.select('h2[class="cs-entry__title"]'):
    print(title_obj)
    print("==========")


```
