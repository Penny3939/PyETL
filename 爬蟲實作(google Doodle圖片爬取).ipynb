### 抓取2023/5月份的圖檔

```python
# 抓取2023/5月份的圖檔

from urllib.request import urlopen
import json
url = "https://www.google.com/doodles/json/2023/5?hl=zh-TW"
response = urlopen(url)

# print(response.read())

doodles = json.load(response)     #將原始資料改成json格式list
for d in doodles:                 #d是dictionary字典
    url = "https:" + d["url"]     #因url沒有前面的https
    print(d["title"],url)         #print出title欄位 &url欄位


# 方法一: 
    #針對圖片做處理
    response = urlopen(url)
    img = response.read()
    fname = "05_doodles/" + url.split("/")[-1]  #要先建立"05_doodles"資料夾; 檔名+取網址的最後一段

    #存檔三部曲 (將圖片抓下來至指定的資料夾裡)
    f = open(fname, "wb")
    f.write(img)
    f.close()

    
# 方法二:記得要import urlretrieve (此方法更快速,但先熟練方法一)
# fname = "0420_doodles/" + url.split("/")[-1]  #前面的"0420_doodles"為資料夾名字+取網址的最後一段
# urlretrieve(url,fname)                        #即可完成下載檔案至資料夾裡
```

 # 抓取2023/ 1~5月份的每月圖檔,每月的資料需獨立一個資料夾

```python

from urllib.request import urlopen, urlretrieve
import json
import os #檢查資料夾是否存在


for month in range(0,6):  #爬取1~5月份 [0,1,2,...,5]
    url = "https://www.google.com/doodles/json/2023/" + str(month+1) +"?hl=zh-TW"
    print("現在處裡的頁面:", url)   #用來檢視月跟月之間
    response = urlopen(url)

    doodles = json.load(response) #將原始資料改成json格式list
    for d in doodles:             #d是dictionary字典
        url = "https:" + d["url"] #因url沒有前面的https
        print(d["title"],url) #print出title欄位 &url欄位

        #要先建立doodles資料夾,下載1-5月的圖文檔至下方dirname指定的資料夾裡
        dirname = "doodles/" + str(month+1) + "/"
        if not os.path.exists(dirname):        #若沒有資料夾,則創建資料夾
            os.mkdir(dirname)

        fname = dirname + url.split("/")[-1]  #前面的"dirname"為資料夾名字+取網址的最後一段
        urlretrieve(url,fname)
```
